```python
#Examples of method in Python

#--Parameterless Method--
print("===================\nThe no_params method takes in NO parameters and just prints out 'I love Chicken Rice!'")
def no_params():
  print('I love Chicken Rice!')

#--Parameterized Functions--
print("===================\nThe with_params method takes ONE parameter and prints out what you like to eat!")
def with_params(food):
  print(f'I love eating {food}')

#--Parameterized Functions WITH Return--
print("===================\nThe addition method takes 2 parameters and returns the sum of those 2")
def addition(a,b):
  c = a + b
  return c
```

# ‚ú®‚ú®‚ú® TRY IT OUT ‚ú®‚ú®‚ú®

```python
#How to RUN a method in Python

#--Parameterless Method--
print("===================\nThis parameterless method takes in NO parameters and just prints out 'I love Chicken Rice!'")
no_params()

#--Parameterized Functions--
print("===================\nThis parameterized method takes ONE parameter and prints out what you like to eat!")
with_params(food = "Cookies")

#--Parameterized Functions WITH Return--
print("===================\nThis parameterized method takes 2 parameters and returns the sum of those 2")
result = addition(1,2); print(result)
```
